```python
import pandas as pd
import glob

csv_files = sorted(glob.glob("simulation_*.csv"))
combined_df = pd.read_csv(csv_files[0])

for file in csv_files[1:]:
    df = pd.read_csv(file)
    combined_df = pd.concat([combined_df, df], ignore_index=True)

combined_df.to_csv("stage_4_1_8wl.csv", index=False)
```
